```python
As implantações poder gerar diferentes status de implantação:
    
    -> error
    -> failure
    -> pending
    -> in_progress
    -> queued
    -> sucess

    
Para mesclar e extrair uma branch é preciso primeiro implantá-lo e verificá-lo.
Esses são os três ambientes possíveis:
    
    -> production
    -> staging
    -> qa
    obs: o ambiente padrão é o production
        
Existem vários parâmetros para o deploy
```

```python
import json
import requests

header = {
    'Authorization': '<token>',
    'Accept': 'application/vnd.github.v3+json'
}

dados = json.dumps({
    'ref': 'master'
})

url = 'https://api.github.com/repos/RenanAlves13/Metodos-Avancados-de-Programacao/deployments'

resposta = requests.post(url, headers=header, data=dados)
print(json.dumps(resposta.json(), indent=4))
```

```python
header = {
    'Authorization': '<token>',
    'Accept': 'application/vnd.github.v3+json'
}

url = 'https://api.github.com/repos/RenanAlves13/Metodos-Avancados-de-Programacao/deployments/693719026'

resposta = requests.delete(url, headers=header)

if resposta.status_code == 204:
    print('O deploy foi deletado com sucesso.')
else:
    print('Algo está errado.')    
```
