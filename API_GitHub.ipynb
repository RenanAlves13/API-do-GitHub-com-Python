```python
get the number of requests an unauthenticated user can make to the GitHub API and the expiration date:
```

```python
import requests

header = {
    'Authorization': 'token'
}

url = 'https://api.github.com/v3/zen'
resposta = requests.get(url, headers=header)

print('Limite: '+resposta.headers['X-RateLimit-Limit'])

if 'GitHub-Authentication-Token-Expiration' not in resposta.headers:
    print('Sem data de experiração')
else:
    print('Data de experiração: '+resposta.headers['GitHub-Authentication-Token-Expiration']) 
```

```python
Create a github repository

--> As respostas podem ser:
    
    *201: O repositório foi criado.
    *403: É proibido que o repositório seja criado
    *422: A validação do código acima falhou.
```

```python
header = {
    'Authorization': 'token',
    'Accept': 'application/vnd.github.v3+json'
}

url = 'https://api.github.com/user/repos'

data = json.dumps({
    'name': 'Estudos sobre a API do GitHub'
})

resposta = requests.post(url, headers=header, data=data)
print(json.dumps(resposta.json(), indent=4))
```

```python
Update a repository

--> A resposta HTTP pode ser:
    
    *200: O repositório foi atualizado com sucesso.
    *307: Houve um redirecionamento temporário.
    *403: É proibido atualizar o repositório.
    *404: O repositório a ser atualizado não foi encontrado.
    *422: A validação do código acima falhou.
```

```python
header = {
    'Authorization': 'token',
    'Accept': 'application/vnd.github.v3+json'
}

data = json.dumps({
    'name': 'API do GitHub'
})

url = "https://api.github.com/repos/RenanAlves13/Estudos-sobre-a-API-do-GitHub"

resposta = requests.patch(url, headers=header, data=data)
print(json.dumps(resposta.json(), indent=4))
```

```python
Delete a repository

--> Pode gerar as seguintes respostas HTTP:
    
    *204: O repositório foi excluído com sucesso.
    *307: Houve um redirecionamento temporário.
    *404: O repositório que estávamos procurando não foi encontrado.
```

```python
headers = {
    'Authorization': 'token',
    'Accept': 'application/vnd.github.v3+json'
}

url = 'https://api.github.com/repos/RenanAlves13/API-do-GitHub'

resposta = requests.delete(url, headers=header)

if resposta.status_code == 204:
    print('O repositório foi deletado com sucesso.')
elif resposta.status_code == 307:
    print('Houve um redirecionamento temporário.')
else:
    print('O repositório não foi encontrado.')
```

```python
Search pull request

OBS:
    
    --> A consulta na linha 6 exibirá os bugs do Python que estão abertos (ou não resolvidos) no Windows. Como a ordem é 
    definida como asc, os bugs mais antigos serão exibidos primeiro.
```
